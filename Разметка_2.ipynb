```python
df = pd.read_csv('/Users/kirillanosov/Downloads/kinopoisk-top250_1.csv', sep = ';')
df
```

```python
df['country'].value_counts()
```

```python
data = df.query('country in ("СССР","Россия")')
```

### Разметка

```python
data['Genres'] = ''

# Визуальный вывод для разметчика
for index, row in data.iterrows():
    print(f"Фильм: {row['movie']}")
    print(f"Описание: {row['overview']}")
    print("Жанры: ")
    genres = input("Введите жанры (через запятую): ")
    data.at[index, 'Genres'] = genres
    print("\n" + "="*30 + "\n")
```

### Итоги
