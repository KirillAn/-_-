## Применение синтетических данных


### Описание кейса

В качестве примера кейса по применению синтетических данных, решил рассмотреть пример [Amazon Alexa](https://alexa.amazon.com/)  - голосового ассистента разработанного Amazon.
Разработчики голосового асситсента столкнулись с нехваткой данных, когда готовили релиз для новых регионов на новых языках.

 Изначально команда располагала типовыми запросами на новых языках - "золотые высказывания" - а также получили некоторые данные через сервис [Cleo](https://www.amazon.com/Amazon-Cleo/dp/B01N5QDE0Y), где пользователи молги помогать тренировать языковые модели. Однако этого оказалось недостаточно, поэтому команда разработки Alexa создала два инструмента, которые бы генерировали синтетические данные для обучения голосового помощника.

 #### Первый инструмент:
 
 Использует технику грамматической индукции - создание грамматических правил из примеров конркетного языка. Анализировались типлвые запросы на языке ("золотые высказывания") и изучались общие семантические и синтаксические шаблоны выражений. На выходе получался набор правил для изменения базовых шаблонных предложений с помощью вставок, удалений и замен слов.

Пример, который показывает как можно моделировать запросы на воспроизведение поп- или рок-музыки, с модификаторами "некоторые" или "больше" или без них, а также диаграмма конечной автоматизации. Вопросительный знак указывает на то, что переменная some_more является необязательной.

Из списка из 50 или 60 "золотых высказываний", инструмент для грамматической индукции может извлечь около 100 правил, которые могут генерировать несколько тысяч обучающих предложений всего за несколько секунд.


 #### Второй инструмент:

 Регулируемое пересэмплирование - методика генерации новых предложений через рекомбинацию слов и фраз из доступных данных. Таким образом дополняется датасет для обучения модели на новом языке.

Предположим, что есть предложения "play Camila Cabello" и "can you play a song by Justin Bieber?", которые размечены таким образом, что "Camila Cabello" и "Justin Bieber" относятся к типу ArtistName. В логике инструмента ArtistName - это тип слота, а "Camila Cabello" и "Justin Bieber" - это значения слота.

Инструмент генерирует дополнительные обучающие примеры, заменяя значения слотов — например, создавая "play Justin Bieber" и "can you play a song by Camila Cabello?". Из базы данных Amazon Music берут имена артистов и названия песен и созданиют тысячи дополнительных обучающих предложений.

Слепая замена значений слотов может привести к не очень хорошим последствиям, поэтому важно понимать, какие значения слотов можно безопасно заменять. Например, в предложениях "play jazz music" и "read detective books", слова "jazz" и "detective" будут помечены как тип слота GenreName. Однако маловероятно, что клиенты попросят Alexa включить "детективную музыку", и использование таких нереалистичных обучающих данных может ухудшить производительность результирующей модели.

Чтобы этого не происходило, стараются представлять запросы клиентов в более абстрактной форме. Например, предложение "play Camila Cabello" может быть закодировано как шаблон "другое + ArtistName".


#### Выводы

Эти инструменты позволили значительно улучшить качество обучения моделей NLU (natural-language-understanding) для новых языков, создавая более разнообразные и реалистичные обучающие данные.
